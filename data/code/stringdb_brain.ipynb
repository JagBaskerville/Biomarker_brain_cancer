```python
# Import packages
import pandas as pd
import numpy as np
```

```python
# Read the brain DEGs data
brain_DEGs = pd.read_csv('resLFC_brain_df.csv', sep=',')
brain_DEGs
```

```python
# Filter P <= 0.01 & logFC > 0.5 quantile
brain_DEGs_filtered = brain_DEGs.loc[(brain_DEGs['padj'] <= 0.01) & (brain_DEGs['log2FoldChange'] >= 2)]
brain_DEGs_filtered
```

```python
brain_top_FC_50 = np.quantile(brain_DEGs_filtered.log2FoldChange, 0.5)
brain_DEGs_filtered = brain_DEGs_filtered.loc[(brain_DEGs_filtered['log2FoldChange'] >= brain_top_FC_50)]
brain_DEGs_filtered
```

```python
brain_DEGs_filtered['Unnamed: 0'] = brain_DEGs_filtered['Unnamed: 0'].str.split('.').str[0]
# Extract Ensembl gene IDs
ensembl_gene_ids = brain_DEGs_filtered['Unnamed: 0'].apply(lambda x: x.split('.')[0])
ensembl_gene_ids
```

```python
from pybiomart import Dataset

dataset = Dataset(name='hsapiens_gene_ensembl',
                  host='http://www.ensembl.org')
# Function to query in batches
def query_in_batches(dataset, ensembl_gene_ids, batch_size=100):
    results = []
    for i in range(0, len(ensembl_gene_ids), batch_size):
        batch = ensembl_gene_ids[i:i + batch_size]
        response = dataset.query(
            attributes=['ensembl_gene_id', 'external_gene_name'],
            filters={'link_ensembl_gene_id': batch.tolist()}
        )
        results.append(response)
    return pd.concat(results)

# Query the dataset in batches
ensembl_to_gene_symbol = query_in_batches(dataset, ensembl_gene_ids)

print(ensembl_to_gene_symbol)
```

```python
ensembl_to_gene_symbol = ensembl_to_gene_symbol[pd.notna(ensembl_to_gene_symbol['Gene name'])]
ensembl_to_gene_symbol.to_csv('ensembl_to_gene_symbol_brain.csv', index=False)
```

```python
# Get gene list
genes = ensembl_to_gene_symbol['Gene name'].tolist()
genes_cleaned = [str(gene) for gene in genes if pd.notnull(gene)]
```

```python
import stringdb
string_ids = stringdb.get_string_ids(genes_cleaned)
enrichment_df = stringdb.get_network(string_ids.queryItem)
enrichment_df
```

```python
# Write the brain ppi network data to csv
enrichment_df = pd.DataFrame(enrichment_df)
enrichment_df.to_csv("brain_network.csv", index=False)
```

```python
## STRING network 
import requests  # python -m pip install requests
from time import sleep
import ipywidgets as widgets
from IPython.display import SVG
string_api_url = "https://version-11-5.string-db.org/api"
output_format = "svg"
method = "network"
request_url = "/".join([string_api_url, output_format, method])
# Combine all proteins into a single string with newline characters in URL encoding (%0D)
protein_list = brain_string_db_top30.node_name.tolist()
combined_proteins = "%0D".join(protein_list)
# Set parameters
params = {
    "identifiers": combined_proteins,  # combined list of proteins
    "species": 9606,  # species NCBI identifier
    "network_flavor": "confidence",  # show confidence links
    "caller_identity": "Final_project_python_R12B48007",  # your app name
    "required_score": 900,  # optional: to filter for high confidence interactions  
}
# Call STRING
response = requests.post(request_url, data=params)
# Save the network to file
file_name = "combined_network.svg"
print(f"Saving interaction network to {file_name}")
with open(file_name, 'wb') as fh:
    fh.write(response.content)
sleep(1)
def display_svg_in_widget(svg_file):
    # Create an output widget to display the SVG
    output_widget = widgets.Output()
    # Display the SVG within the output widget
    with output_widget:
        display(SVG(filename=svg_file))
    return output_widget
svg_file = "combined_network.svg"
output_widget = display_svg_in_widget(svg_file)
output_widget
```
